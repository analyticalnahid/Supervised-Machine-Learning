```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train.csv', usecols=['Parch','SibSp','Pclass','Survived'])
df['Family'] = df['Parch']+df['SibSp']
df.drop(columns=['Parch','SibSp'],inplace=True)
df = df[['Pclass','Family','Survived']]
df.head()
```

```python
import plotly.express as px

fig = px.scatter_3d(df,x="Pclass",y="Family",z="Survived")
fig.show()
```

```python
X = df.drop(columns='Survived')
y = df['Survived']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=2)
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(multi_class='ovr')
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt


plot_decision_regions(X=X.values, y=y.values, clf=lr, legend=2)

# Adding axes annotations
plt.xlabel('PClass')
plt.xlabel('Family')
plt.title('Sigmoid on Iris')

plt.show()
```
