### Removing outliners for sepal width

```python
iqr = data['Sepal Width'].quantile(0.75) - data['Sepal Width'].quantile(0.25) 
print('iqr = ',iqr)
up = data['Sepal Width'].quantile(0.75) + 1.5 * iqr
down = data['Sepal Width'].quantile(0.25) - 1.5 * iqr
print('up = {} , down = {}'.format(up,down))
sepalw = data['Sepal Width']

outlier = []
for i in sepalw:
    if i > up:
        print('outliner up ',i)
        outlier.append(i)

for i in sepalw:
    if i < down:
        print('outliner down ',i)
        outlier.append(i)
        

```

```python
data['Sepal Width'] = data['Sepal Width'].replace(outlier,data['Sepal Width'].mean())
```
